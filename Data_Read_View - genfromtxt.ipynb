```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
import datetime
```

```python
f = np.genfromtxt('Grant_0408_0414.txt', skip_header=4)
f2 = np.genfromtxt('Grant_0408_0414.txt', skip_header=4,usecols=[0,1],dtype=str) #date and time
```

```python
timed = f2[:,0] # date
timeh = f2[:,1] # hour:minute
RN = f[:,2]
batt_volts = f[:,3]
ptemp_C = f[:,4]
O3_ppm = f[:,5]
NO2_ppm = f[:,6]
SO2_ppm = f[:,7]
CO_ppm = f[:,8]
VOC_ppm = f[:,9]
NMHC_ppm = f[:,10]
temp_C = f[:,11]
RH = f[:,12]
wspd1 = f[:,13]
wdir1 = f[:,14]
wspd2 = f[:,15]
flx_down = f[:,16]
tot_down = f[:,17]
flx_up = f[:,18]
tot_up = f[:,19]
wvect_spd = f[:,20]
wvect_dir = f[:,21]
#time2 = f[:,22]
```

```python
time = (RN*0)
for i in range(0,time.size):
    if i < (time.size - 1): #total number of values - 1
        time[i+1] = time[i] + 1
```

```python
timenew = time[::320]
timehnew = timeh[::320]
```

```python
plt.figure(figsize=[16,14])

ax1 = plt.subplot(6,1,1)
plt.xticks(timenew, timehnew)
ax1.set_ylim(0,max(O3_ppm))
ax1.plot(O3_ppm, linewidth=0.8, c='r')
ax1.set_ylabel('O3 (ppm)', fontsize=14)

ax2 = plt.subplot(6,1,2)
plt.xticks(timenew, timehnew)
ax2.set_ylim(0,max(NO2_ppm))
ax2.plot(NO2_ppm, linewidth=0.8, c='b')
ax2.set_ylabel('NO2 (ppm)', fontsize=14)

ax3 = plt.subplot(6,1,3)
plt.xticks(timenew, timehnew)
ax3.set_ylim(0,0.3)
ax3.plot(SO2_ppm, linewidth=0.8, c='g')
ax3.set_ylabel('SO2 (ppm)', fontsize=14)

ax4 = plt.subplot(6,1,4)
plt.xticks(timenew, timehnew)
ax4.set_ylim(0,2)
ax4.plot(CO_ppm, linewidth=0.8, c='purple')
ax4.set_ylabel('CO (ppm)', fontsize=14)

ax5 = plt.subplot(6,1,5)
plt.xticks(timenew, timehnew)
ax5.set_ylim(0,1.5)
ax5.plot(VOC_ppm, linewidth=0.8, c='orange')
ax5.set_ylabel('VOC (ppm)', fontsize=14)

ax6 = plt.subplot(6,1,6)
plt.xticks(timenew, timehnew)
ax6.set_ylim(0,1.5)
ax6.plot(NMHC_ppm, linewidth=0.8, c='black')
ax6.set_ylabel('NMHC (ppm)', fontsize=14)

plt.show()
```

```python
plt.figure(figsize=[16,16])

ax1 = plt.subplot(7,1,1)
plt.xticks(timenew, timehnew)
ax1.set_ylim(0,max(temp_C))
ax1.plot(temp_C, linewidth=0.8, c='r')
ax1.set_ylabel('Temp ($^\circ$C)', fontsize=14)

ax2 = plt.subplot(7,1,2)
plt.xticks(timenew, timehnew)
ax2.set_ylim(0,max(RH))
ax2.plot(RH, linewidth=0.8, c='b')
ax2.set_ylabel('RH (%)', fontsize=14)

ax3 = plt.subplot(7,1,3)
plt.xticks(timenew, timehnew)
ax3.set_ylim(0,max(wspd1))
ax3.plot(wspd1, linewidth=0.8, c='g')
ax3.set_ylabel('Wspeed1 (m/s)', fontsize=14)

ax4 = plt.subplot(7,1,4)
plt.xticks(timenew, timehnew)
ax4.set_ylim(0,max(wdir1))
ax4.plot(wdir1, linewidth=0.8, c='purple')
ax4.set_ylabel('Wdirection ($^\circ$)', fontsize=14)

ax5 = plt.subplot(7,1,5)
plt.xticks(timenew, timehnew)
ax5.set_ylim(0,max(flx_up))
ax5.plot(flx_up, linewidth=0.8, c='orange')
ax5.set_ylabel('Solar Atm', fontsize=14)

ax6 = plt.subplot(7,1,6)
plt.xticks(timenew, timehnew)
ax6.set_ylim(0,max(flx_down))
ax6.plot(flx_down, linewidth=0.8, color='black')
ax6.set_ylabel('Solar Ground', fontsize=14)

ax7 = plt.subplot(7,1,7)
plt.xticks(timenew, timehnew)
ax7.set_ylim(0,max(wspd2))
ax7.plot(wspd2, linewidth=0.8, c='limegreen')
ax7.set_ylabel('Wspeed2 (m/s)', fontsize=14)

plt.show()
```
